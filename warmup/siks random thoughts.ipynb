```python
# http://stackoverflow.com/questions/21420792/exponential-curve-fitting-in-scipy
#http://en.wikipedia.org/wiki/Gamma_distribution
#http://en.wikipedia.org/wiki/Stirling%27s_approximation
#http://en.wikipedia.org/wiki/Gamma_function
def myGamma(x, k, b):
    def Tau(n):
        if n < 0:
            return 1.
        else:
            return ((2.*np.pi*n)**.5) * ((n/np.exp(1))**n)
    return ((x**(k-1.)) * np.exp(-x/b)) / ((b**k)*Tau(k-1.))

#plt.subplot(nrows=1,ncols=1,figsize=(20,10))
x = np.linspace(0, 20, 100)

for k, O, currentColor in zip([1, 2, 3, 5, 9, 7.5, 0.5], 
                              [2, 2, 2, 1, 0.5, 1, 1], 
                              sns.color_palette("hls", 7)):
    #print 'calculating k:{0:2f} O:{1:2f}'.format(k,O)
    yy = [myGamma(xx, k, O) for xx in x]
    plt.plot(x, yy, c=currentColor, label='k:{0:0.2f}   O:{1:0.2f}'.format(k,O))
plt.axis([0, 20, 0, .5])

plt.legend()
plt.show()
```

```python
from scipy.optimize import curve_fit
# Notice that using the default initialization k=1., O=1. the fitting method would fail check pcov everything is inf.
# Therefore start the initial values at some different place. Pick a distribution from the previous figure.
popt, pcov = curve_fit(myGamma, data.index, data['run00'].get_values(), p0=(2.,2.))
print popt
print pcov

fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(20,10))
ax[0].scatter(data.index, data['run00'].get_values(), alpha=.5)
x = np.linspace(0, 500, 300)
y = [myGamma(xx, popt[0], popt[1]) for xx in x]
ax[0].plot(x,y,'r-', label='k:{0:0.2f}   O:{1:0.2f}'.format(k,O))
ax[0].set_title('NO data zero-lining')

# Get the convergence value using all the runs to zeroline the data
zeroLine = data[-10:].get_values().mean()
# Recompute the parameters
popt, pcov = curve_fit(myGamma, data.index, data['run00'].get_values()-zeroLine, p0=(2.,2.))
ax[1].scatter(data.index, data['run00'].get_values()-zeroLine, alpha=.5)
x = np.linspace(0, 500, 300)
y = [myGamma(xx, popt[0], popt[1]) for xx in x]
ax[1].plot(x,y,'r-')
ax[1].set_title('With zero-lining')

plt.show()
```

##Outliers treatmend

In my opinnion, the function should be robust to outliers. You want to fit the curve you can actually see. Since the used minimization is Least Squares Minimization, the measure is not robust to noice. Our proposal would be to use the same principal as robust estimation using ransac. We outcast the outliers and refine the fitting.

However instead of using ransac we'll do it straight.

```python
#dataError=data.apply(lambda row: row., axis=1)
data = data
data['xx'] = data.index
dataError = data.apply(lambda row: row[:-1] - myGamma(row[-1],firstParam[0], firstParam[1]), axis=1)
dataErrorX = np.asarray(dataError.index).repeat(len(dataError.columns))
dataErrorY = np.concatenate(dataError.get_values())

data = data.drop('xx', 1)
outlierTh = dataY.std()

## prepare the axis to plot
fig = plt.figure(figsize=(20,10))
ax1 = plt.subplot2grid((3,4), (1,0), colspan=3, rowspan=3)
ax2 = plt.subplot2grid((3,4), (1, 3), rowspan=3)
plt.tight_layout()

# plot
ax1.hexbin(dataErrorX,dataErrorY,gridsize = 50, cmap=plt.cm.Blues)
sns.kdeplot(dataErrorY,ax=ax2, bw=0.0005, vertical=True, shade=True)
ax2.set_ylim(ax1.get_ylim())
for myAx in (ax1, ax2):
    myAx.plot(myAx.get_xlim(),[outlierTh, outlierTh], 'r:')
    myAx.plot(myAx.get_xlim(),[-outlierTh, -outlierTh], 'r:')
plt.show()
```
